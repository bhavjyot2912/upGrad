```python
def walker(video_url):
    content = req.get(video_url)
    soup = BeautifulSoup(content.content, 'html.parser')
    for script in soup.findAll('script'):
        if 'keywords' in str(script):
            start_ind = str(script).index('keywords')
            start_ind += 10
            rest_out = str(str(script)[start_ind:])
            end_ind = rest_out.index(']')
            final = rest_out[:end_ind + 1]
            final = final.replace('\\', '')
            final = final.replace(':', '')
            return json.loads(final)

results_dict = {}

for link in youtube_links_list_simplilearn:
    tags = walker(link)
    results_dict[link] = tags

print("Results:")
for link, tags in results_dict.items():
    print(f"Link: {link}, Tags: {tags}")
    
# video_url = input("Enter the URL of the video: ")
# tags = walker(video_url)
# print("Tags:", tags)
```

```python
with open('sample_simplilearn.txt', 'w') as file:
    json.dump(results_dict, file, indent=2)
```
